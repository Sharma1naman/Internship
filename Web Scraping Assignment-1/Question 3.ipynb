# 3) Write a python program to display IMDB’s Top rated 50 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send a GET request to IMDb's top rated Indian movies page
url = "https://www.imdb.com/india/top-rated-indian-movies/"
response = requests.get(url)
```

```python
# Use BeautifulSoup to parse the HTML content of the response
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the table that contains the top 50 rated Indian movies
table = soup.find("tbody", {"class": "lister-list"})
```

```python
# Extract the data for each movie in the table
movies = []
for row in table.find_all("tr")[0:50]:
    movie = {}
    movie["Name"] = row.find("td", {"class": "titleColumn"}).a.text
    movie["Rating"] = float(row.find("strong").text)
    movie["Year of release"] = int(row.find("span", {"class": "secondaryInfo"}).text[1:5])
    movies.append(movie)
```

```python
# Create a Pandas DataFrame to store the results
df = pd.DataFrame(movies)
```
