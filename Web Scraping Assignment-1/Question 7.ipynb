# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame- i) Headline ii) Time iii) NewsLink

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
# Send a GET request to the website
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
```

```python
# Use BeautifulSoup to parse the HTML content of the response
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Extract the headline data for each news
headline = []

for i in soup.find_all('a', class_ = "LatestNews-headline" ):
    headline.append(i.text)
    
headline
```

```python
# Extract the time data for each news
time = []

for i in soup.find_all('time', class_= "LatestNews-timestamp"):
    time.append(i.text)
    
time
```

```python
# Extract the NewsLink data for each news
NewsLink= []

for i in soup.find_all('a', class_="LatestNews-headline"):
    NewsLink.append(i.get('href'))
    
NewsLink
```

```python
# printing length
print(len(headline), len(time), len(NewsLink))
```

```python
# Create a Pandas DataFrame to store the results
df = pd.DataFrame({'Headline':headline, 'Time':time, 'News Link':NewsLink})
```
